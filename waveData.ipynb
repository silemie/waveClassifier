```python
d3 = pd.read_table('data3.dat', sep = ',', header = None)
d3 = d3[[2,3,4,5,6,7,8,9]]
d3 = d3[:100]
```

```python
plt.plot(d3)
plt.show()
```

```python
data = np.fromfile('data2.dat', dtype = float, sep = ',')
mfccs = librosa.feature.mfcc(y=data)
mfccsscaled = np.mean(mfccs.T,axis=0)
plt.plot(mfccs)
plt.show()
```

```python
plt.figure(figsize=(10, 4))
librosa.display.specshow(mfccs, x_axis='time')
plt.colorbar()
plt.title('MFCC')
plt.tight_layout()
plt.show()
```

```python
import os 

path = '/Volumes/Untitled/LBData/异常波形/'
#path = '/Users/miezai/Desktop/'
os.chdir(path)
file_chdir = os.getcwd()

filelist = []

for root, dirs, files in os.walk(file_chdir):
    for file in files:
        if os.path.splitext(file)[1] == '.dat' or os.path.splitext(file)[1] == '.Dat' or os.path.splitext(file)[1] == '.DAT':
            filelist.append(file)

data = pd.DataFrame()
count = 0

for dat in filelist:
    try:
        d = pd.read_table(dat, header = None, sep = ',')
        d = d[[2,3,4,5,6,7,8,9,10]]
        d = d[:100]
        d = d.T
        data = data.append(d)
        count = count + 1
        print("Successfully load ", dat)
    except:
        continue

print("Successfully load ", count)
```
